```python
import networkx as nx
import pandas as pd
import matplotlib.pyplot as plt
```

```python
nodes = pd.read_csv('F:\FCIS\socialNetwork_Project\metadata_primaryschool_Nodes.csv')
```

```python
g = nx.Graph()
for x in range(len(nodes['ID'])):
    g.add_node(nodes['ID'][x])
    
```

```python
g.nodes()
```

```python
edges = pd.read_csv('F:\FCIS\socialNetwork_Project\primaryschool_Edges.csv')
```

```python
edges
```

```python
for x in range(len(edges['Source'])):
    g.add_edge(edges['Source'][x], edges['Target'][x])
```

```python
nx.draw(g)
```

```python
plt.figure(figsize=(10,10))
nx.draw_networkx(g)
plt.show()
```

```python
import PySimpleGUI as sg
import matplotlib.pyplot as plt
from matplotlib.backends.backend_tkagg import FigureCanvasTkAgg
   
  
def create_bar_graph():
    plt.figure(figsize =(10, 7))
    nx.draw_networkx(g)
    return plt.gcf()

layout = [[sg.Text('Bar Graph')],
          [sg.Canvas(size=(900, 900), key='-CANVAS-')],
          [sg.Exit()]]

def draw_figure(canvas, figure):
    figure_canvas_agg = FigureCanvasTkAgg(figure, canvas)
    figure_canvas_agg.draw()
    figure_canvas_agg.get_tk_widget().pack(side='top', fill='both', expand=1)
    return figure_canvas_agg


window = sg.Window('PySimpleGUI + MatPlotLib Bar Graphs', layout, finalize=True, element_justification='center')

draw_figure(window['-CANVAS-'].TKCanvas, create_bar_graph())

while True:
    event, values = window.read()
    if event == sg.WIN_CLOSED or event == 'Exit':
        break

window.close()
```
